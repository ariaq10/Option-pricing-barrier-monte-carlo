## Imports

This block imports the main libraries used in the notebook:

- `numpy` for numerical work and simulation,
- `matplotlib` for plotting,
- `scipy.stats.norm` for the normal distribution functions that appear in the Black-Scholes formula.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Parameter setup, GBM simulation, and option pricing

This block defines the model parameters for the stock and option, sets the time discretisation, and simulates sample paths of a geometric Brownian motion under the risk-neutral measure.

After generating the paths, it computes:

- the terminal payoff of the European call option,
- the Monte Carlo estimate of the option price,
- the closed-form Black-Scholes price for comparison.


```python
# Parameters
S0 = 100        # initial stock price
K = 100         # strike price
T = 1.0         # time to maturity (1 year)
r = 0.05        # risk-free rate
mu = 0.1        # drift (not used in risk-neutral valuation)
sigma = 0.2     # volatility
n_steps = 252   # daily steps
n_paths = 10_000 # number of simulated paths

dt = T / n_steps

# Simulate GBM paths
np.random.seed(42)
Z = np.random.standard_normal((n_paths, n_steps))
S = np.zeros((n_paths, n_steps + 1))
S[:, 0] = S0

for t in range(1, n_steps + 1):
    S[:, t] = S[:, t-1] * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z[:, t-1])

# Monte Carlo option price
payoff = np.maximum(S[:, -1] - K, 0)
mc_price = np.exp(-r * T) * np.mean(payoff)

# Black-Scholes analytic price
d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)

bs_price = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
```

## Numerical output

This block prints the two prices obtained above so we can compare the Monte Carlo estimate with the Black-Scholes benchmark.


```python
# Output
print(f"Monte Carlo European Call Price: {mc_price:.4f}")
print(f"Black-Scholes European Call Price: {bs_price:.4f}")
```

## Plot of simulated GBM paths

This block plots a sample of the simulated stock-price paths over time, giving a visual picture of how the geometric Brownian motion evolves.


```python
# Plot sample paths
plt.figure(figsize=(15, 6))
for i in range(20):
    plt.plot(np.linspace(0, T, n_steps + 1), S[i])
plt.title("Simulated Geometric Brownian Motion Paths")
plt.xlabel("Time (years)")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```

## ITM/OTM path visualisation

This block plots some simulated paths again, but now highlights whether the option is in the money or out of the money along each path:

- red segments correspond to `S_t \geq K`,
- blue segments correspond to `S_t < K`.

The dashed horizontal line marks the strike price.


```python

# Plotting GBM paths with ITM / OTM coloring
time_grid = np.linspace(0, T, n_steps + 1)
plt.figure(figsize=(12,7))

for i in range(20): #or put n_paths to see all paths but less clear
    S_i = S[i]
    # Identify regions
    itm_mask = S_i >= K
    otm_mask = S_i < K
    # Plot ITM parts
    plt.plot(time_grid[itm_mask], S_i[itm_mask], color='red', lw=1)
    # Plot OTM parts
    plt.plot(time_grid[otm_mask], S_i[otm_mask], color='blue', lw=1)

plt.axhline(K, color='black', linestyle='dashed', label=f'Strike Price = {K}')
plt.title("Simulated GBM Paths: Red=In the Money (ITM), Blue=Out of the Money (OTM)")
plt.xlabel("Time (years)")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

## Payoff distribution plots

This block studies the simulated payoffs at maturity.

It first plots a histogram of the discounted payoffs and overlays the Monte Carlo price and Black-Scholes price.  
It then shows a scatter plot of the raw payoff for each simulated path, which helps visualise how many paths finish with zero payoff versus a positive payoff.


```python
discounted_payoff = np.exp(-r * T) * payoff

# Plot 1: Distribution of simulated discounted option prices
plt.figure(figsize=(10,6))
plt.hist(discounted_payoff, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(bs_price, color='red', linestyle='dashed', linewidth=2, label=f'Black-Scholes Price = {bs_price:.2f}')
plt.axvline(mc_price, color='blue', linestyle='dashed', linewidth=2, label=f'MC Price = {mc_price:.2f}')
plt.title("Distribution of Simulated Discounted Option Payoffs")
plt.xlabel("Discounted Payoff")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()


# Plot 2: Payoff for each path (scatter plot)
plt.figure(figsize=(10,6))
plt.scatter(range(len(payoff)), payoff, alpha=0.3, color='purple', s=5)
plt.title("European Call Payoff at Maturity for Each Path")
plt.xlabel("Path index")
plt.ylabel("Payoff")
plt.grid(True)
plt.show()
```

## Sensitivity plots from the Black-Scholes formula

This block defines a Black-Scholes call-pricing function and then varies one input at a time while keeping the others fixed.

The resulting plots show how the call option price changes with:

- the initial stock price `S0`,
- the strike `K`,
- the maturity `T`,
- the volatility `sigma`,
- the risk-free rate `r`.

These graphs are useful for building intuition about the qualitative behaviour of European call prices.


```python
#Here we used the black scholes formular to plot the relationship between the call-price and the other inputs, assuming the rest are fixed.


def black_scholes_call(S0, K, T, r, sigma):
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# Fixed parameters
S0 = 100
K = 100
T = 1.0
r = 0.05
sigma = 0.2

# Plot 1: Price vs S0
S0_range = np.linspace(80, 120, 100)
prices_S0 = [black_scholes_call(S, K, T, r, sigma) for S in S0_range]
plt.figure(figsize=(10,6))
plt.plot(S0_range, prices_S0)
plt.title("Call Price vs Spot Price (S0)")
plt.xlabel("Spot Price S0")
plt.ylabel("Call Price")
plt.grid(True)
plt.show()

# Plot 2: Price vs K
K_range = np.linspace(80, 120, 100)
prices_K = [black_scholes_call(S0, K_, T, r, sigma) for K_ in K_range]
plt.figure(figsize=(10,6))
plt.plot(K_range, prices_K)
plt.axvline(K, color='grey', linestyle='dotted', label=f'Original Strike = {K}')
plt.title("Call Price vs Strike Price (K)")
plt.xlabel("Strike Price K")
plt.ylabel("Call Price")
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Price vs T
T_range = np.linspace(0.01, 2, 100)
prices_T = [black_scholes_call(S0, K, T_, r, sigma) for T_ in T_range]
plt.figure(figsize=(10,6))
plt.plot(T_range, prices_T)
plt.title("Call Price vs Maturity (T)")
plt.xlabel("Maturity T (years)")
plt.ylabel("Call Price")
plt.grid(True)
plt.show()

# Plot 4: Price vs sigma
sigma_range = np.linspace(0.01, 1.0, 100)
prices_sigma = [black_scholes_call(S0, K, T, r, sigma_) for sigma_ in sigma_range]
plt.figure(figsize=(10,6))
plt.plot(sigma_range, prices_sigma)
plt.title("Call Price vs Volatility (sigma)")
plt.xlabel("Volatility sigma")
plt.ylabel("Call Price")
plt.grid(True)
plt.show()

# Plot 5: Price vs r
r_range = np.linspace(0.0, 0.2, 100)
prices_r = [black_scholes_call(S0, K, T, r_, sigma) for r_ in r_range]
plt.figure(figsize=(10,6))
plt.plot(r_range, prices_r)
plt.title("Call Price vs Risk-Free Rate (r)")
plt.xlabel("Risk-Free Rate r")
plt.ylabel("Call Price")
plt.grid(True)
plt.show()
```
